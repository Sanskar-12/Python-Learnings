```python
def even_num(num):
    if num%2==0:
        return True
    else:
        return False
        
```

```python
even_num(2)
```

```python
even_num(3)
```

```python
lst=[1,2,3,4,5,6,7,8,9]

```

```python
def evenoddsum(lst):
    even_sum=0
    odd_sum=0
    for i in lst:
        if i%2==0:
            even_sum=even_sum+i
        else:
            odd_sum=odd_sum+i

    return even_sum,odd_sum
```

```python
evenoddsum(lst)
```
