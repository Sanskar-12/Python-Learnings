## String Formatting


```python
print("Hello world")
```

```python
def greet(name):
    return "Hello {}. Welcome to the community".format(name)
```

```python
greet("Sanskar")
```

```python
def greet1(name,age):
    return "Hello {name} and age {age}. Welcome to the community".format(age=age,name=name)
```

```python
greet1("Sanskar",10)
```
