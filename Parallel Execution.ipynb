```python
import time
from concurrent.futures import ThreadPoolExecutor

def returnnumber(a):
    time.sleep(1)
    return a
```

```python
start=time.time()
with ThreadPoolExecutor(max_workers=2) as executor: # max workers=2 means two thread will be created and will be executing parallely
    for result in executor.map(returnnumber,range(30)):
        print("Count: {}".format(result))
print("The total time is {}".format(time.time()-start))
```
