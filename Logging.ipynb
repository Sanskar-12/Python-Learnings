```python
import logging
```

```python
logging.basicConfig(
    filename='Logs.txt',
    filemode='a',
    format="%(asctime)s - %(levelname)s - %(message)s",
    datefmt='%Y-%m-%d %H-%M-%S'
)
```

```python
for i in range(15):
    if(i==3):
        logging.warning("Log warning message")
    if(i==5):
        logging.critical("Log critical message")
    if(i==10):
        logging.error("Log error message")
```
