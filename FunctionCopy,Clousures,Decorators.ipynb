## Closures

```python
def main_welcome():
    msg="Hello"
    def welcome():
        print("Hello 1")
        print(msg) # we can access this msg function inside also is called closure
        print("Hello 2")
    return welcome()
```

```python
main_welcome()
```

```python
# A closure is a function that remembers variables from its outer function, even after the outer function has finished executing.
```

```python
# Decorators (passing function inside a function)

def main_welcome(func):
    def welcome():
        print("Hello 1")
        func("Hello") 
        print("Hello 2")
    return welcome()
```

```python
main_welcome(print)
```

```python
def main_welcome(func): # passing channel_name inside main_welcome
    def welcome():
        print("Hello 1")
        func()
        print("Hello 2")
    return welcome()
```

```python
@main_welcome
def channel_name():
    print("Hello")
```

```python
main_welcome(channel_name)
```
