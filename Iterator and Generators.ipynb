## Iterators

```python
# Iterator is used to iterate elements inside list
```

```python
lst=[1,2,3,4]
```

```python
for i in lst:
    print(i)
```

```python
iterable=iter(lst)
```

```python
for i in iterable:
    print(i)
```

```python
next(iterable)
```

```python
# Generator is used to create iterator
```

```python
def square(n):
    for i in range(n):
        yield i**2     # used to generate an iterator
```

```python
for i in square(3):
    print(i)
```
