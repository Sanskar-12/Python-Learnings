```python
from PyPDF2 import PdfReader,PdfWriter
```

```python
# read the pdf file
file_pdf=PdfReader('sanskar_resume.pdf')
# object for writing in pdf file
out_pdf=PdfWriter()
```

```python
len(file_pdf.pages)
```

```python
for i in range(len(file_pdf.pages)):
    # get the page details from the input pdf
    pageDetails=file_pdf.pages[i]
    # add the page into the output pdf
    out_pdf.add_page(pageDetails)
```

```python
password="Sanskar@123"
out_pdf.encrypt(password)
```

```python
with open("encryptedresume.pdf","wb") as filename:
    out_pdf.write(filename)
```
