## Python Exception Handling

```python
try:
    # code block where exception can occur
    a=b
except NameError as ex1: # Custom Exception
    print("User has not defined the variable")
except Exception as ex:
    print(ex)
```

```python
try:
    a=1
    b='s'
    c=a+b
except TypeError:
    print("Try to make the datatype similar")
except NameError: # Custom Exception
    print("User has not defined the variable")
except Exception as ex:
    print(ex)
```

```python
try:
    a=int(input("Enter the number 1 "))
    b=int(input("Enter the number 2 "))
    c=a/b
except ZeroDivisionError:
    print("Provide number greater than 0")
except TypeError:
    print("Try to make the datatype similar")
except NameError: # Custom Exception
    print("User has not defined the variable")
except Exception as ex:
    print(ex)
```

```python
# try else finally

try:
    a=int(input("Enter the number 1 "))
    b=int(input("Enter the number 2 "))
    c=a/b
except ZeroDivisionError:
    print("Provide number greater than 0")
except TypeError:
    print("Try to make the datatype similar")
except NameError: # Custom Exception
    print("User has not defined the variable")
except Exception as ex:
    print(ex)
else:
    print(c)
finally:
    print("The code has ended")
```

## Custom Exception

```python
class Error(Exception):  # Error class inherits Exception class
    pass
```

```python
class dobException(Error): # dobException inherits Error class (Custom Exception)
    pass
    
```

```python
def isEligible(year):
    age=2026-year
    try:
        if age<=30 and age>20:
            return True
        else:
            raise dobException # raising the custom exception
    except dobException:
        print("Age must be more than 20 and less than or equal to 30 years")
```

```python
isEligible(int(input("Enter the year of birth ")))
```
