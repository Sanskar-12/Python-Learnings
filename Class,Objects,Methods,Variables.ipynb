```python
# This is how to create a class 

class Car:
    def __init__(self,window,wheel,enginetype):
        self.windows=window
        self.wheels=wheel
        self.enginetypes=enginetype

    def self_driving(self):
        return "This is a {} car".format(self.enginetypes)
```

```python
car1=Car(4,4,"petrol")
```

```python
print(car1.self_driving())
```

```python
print(car1.windows)
```
