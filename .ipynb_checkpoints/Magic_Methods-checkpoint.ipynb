```python
class Car:
    def __init__(self,window,wheel,enginetype):
        self.windows=window
        self.wheels=wheel
        self.enginetypes=enginetype
    def __str__(self):                    # overriding str method
        return "obj has been initialised"
    def self_driving(self):
        return "This is a {} car".format(self.enginetypes)
```

```python
c=Car(2,2,"petrol")
```
