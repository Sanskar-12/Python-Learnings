## Lambda Functions

### Functions with no name (anonymous function)

```python
addition=lambda a,b:a+b
```

```python
addition(1,2)
```

## Map Functions used to Iterate 

```python
def even_odd(num):
    if num%2==0:
        return True
    else:
        return False
```

```python
lst=[1,2,3,4,5]
```

```python
list(map(even_odd,lst))
```

## Filter Function used to filter values

```python
def even(num):
    if num%2==0:
        return True
```

```python
list(filter(even,lst))
```

```python
list(filter(lambda num:num%2==0,lst))
```
