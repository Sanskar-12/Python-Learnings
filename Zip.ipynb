zip() is used to combine multiple iterables (lists, tuples, etc.) element-wise.

```python
lst1=[1,2,3]
lst2=['a','b','c']
```

```python
output=zip(lst1,lst2)
```

```python
list(output)
```

```python
next(output)
```

```python
output=zip(lst1,lst2)

for i,j in output:
    print(i)
    print(j)
```

```python
dict1={"a":1,'b':2,'c':3}
dict2={'d':4,'e':5,'f':6}
```

```python
output=zip(dict1.items(),dict2.items())
```

```python
for (i,j),(i2,j2) in output:
    print(i,j)
    print(i2,j2)
```
