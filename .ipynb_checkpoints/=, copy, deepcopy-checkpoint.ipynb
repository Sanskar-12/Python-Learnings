```python
# '=' operator
```

```python
lst1=[1,2,3,4]
lst2=lst1
```

```python
lst1[0]=100
```

```python
lst1, lst2
```

```python
# copy operator (shallow copy)
```

```python
lst1=[1,2,3,4]
lst2=lst1.copy()
```

```python
lst1[0]=100
```

```python
lst1,lst2
```

```python
# shallow copy of nested list

lst1=[[1,2,3,4],[5,6,7,8]]
lst2=lst1.copy()
```

```python
lst1[0][0]=100
```

```python
lst1,lst2
```

```python
lst1.append([12,23,445,65])
```

```python
lst1,lst2
```

```python
# deep copy of nested list
import copy

lst1=[[1,2,3,4],[5,6,7,8]]
lst2=copy.deepcopy(lst1)
```

```python
lst1[0][0]=100
```

```python
lst1,lst2
```
