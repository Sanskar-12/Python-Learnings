# Multiple Inheritance


```python
class A:
    def method1(self):
        print("A class method1 is called")
```

```python
class B(A):
    def method1(self):
        print("B class method1 is called")
```

```python
class C(A):
    def method1(self):
        print("C class method1 is called")
```

```python
class D(B,C):
    def method1(self):
        print("D class method1 is called")
```

```python
d=D()
d.method1()
```

```python
C.method1(d) # calling C method1 from d object
```

```python
A.method1(d) # calling A method1 from d object
```
