## Dictionaries

```python
dic={}
```

```python
type(dic)
```

```python
dic={"Car":"BMW","Bike":"Honda"}
```

```python
type(dic)
```

```python
## Access the item in the dic

dic["Car"]
```

```python
## for in dict

for x in dic:
    print(x)
```

```python
for x in dic.values():
    print(x)
```

```python
for x in dic.items():
    print(x)
```

```python
## Add in dic

dic["Car2"]="Maruti"

dic
```

## Nested Dictionary

```python
car1_model={'Mercedes':1960}
car2_model={'Audi':1958}
car3_model={'Ferrari':2323}
```

```python
car={'car1':car1_model,'car2':car2_model,'car3':car3_model}

car
```

```python
car['car1']
```

```python
car['car1']['Mercedes']
```
